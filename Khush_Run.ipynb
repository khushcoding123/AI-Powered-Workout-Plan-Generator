```python
import pandas as pd

all_data = pd.concat([
    pd.read_csv('UpdatedActivity.csv'),
    pd.read_csv('UpdatedDailyActivity.csv'),
    pd.read_csv('UpdatedExerciseDataset.csv'),
    pd.read_csv('UpdatedGymTracking.csv'),
    pd.read_csv('UpdatedWorkoutTracker.csv')
], ignore_index=True)

# Delete rows with zero duration
all_data = all_data[all_data['exercise_duration'] != 0]  # <-- NEW LINE

all_data.to_csv('Combined_Fitness_Data.csv', index=False)

# Ensure only valid exercise_type/exercise_name pairs remain
exercise_type_rules = {
    'Aerobic': ['Cycling', 'Running', 'Swimming', 'Dancing', 'Rowing'],
    'Anaerobic': ['HIIT', 'Kickboxing'],
    'Strength': ['Weightlifting'],
    'Flexibility': ['Yoga', 'Pilates']
}
def valid_type(row):
    return row['exercise_name'] in exercise_type_rules.get(row['exercise_type'], [])
all_data = all_data[all_data.apply(valid_type, axis=1)]
all_data.to_csv('Combined_Fitness_Data.csv', index=False)

# Numeric encoding for training
df = all_data.copy()
exercise_name_mapping = {name: idx+1 for idx, name in enumerate(df['exercise_name'].unique())}
exercise_type_mapping = {name: idx+1 for idx, name in enumerate(df['exercise_type'].unique())}
gender_mapping = {'Female': 1, 'Male': 2, 'Other': 3}
df['exercise_name'] = df['exercise_name'].map(exercise_name_mapping).fillna(0).astype(int)
df['exercise_type'] = df['exercise_type'].map(exercise_type_mapping).fillna(0).astype(int)
df['gender'] = df['gender'].map(gender_mapping).fillna(0).astype(int)
df = df.fillna(0)
df.to_csv('Combined_Fitness_Data_Numerical.csv', index=False)
```

```python
%%writefile app.py
import pandas as pd
import streamlit as st
import tensorflow as tf
import numpy as np
from joblib import load

model = tf.keras.models.load_model('workout_recommender.keras')
scaler = load('scaler.joblib')
encoder = load('encoder.joblib')
classes = np.load('class_labels.npy', allow_pickle=True)

df = pd.read_csv('Combined_Fitness_Data.csv')
duration_data = df.groupby(['exercise_type', 'exercise_name'])['exercise_duration'].median().reset_index()

exercise_type_rules = {
    'Aerobic': ['Cycling', 'Running', 'Swimming', 'Dancing', 'Rowing'],
    'Anaerobic': ['HIIT', 'Kickboxing'],
    'Strength': ['Weightlifting'],
    'Flexibility': ['Yoga', 'Pilates']
}
gender_mapping = {'Female': '1', 'Male': '2', 'Other': '3'}

st.title('FitScout AI Workout Recommender')

with st.form("user_inputs"):
    age = st.number_input('Age', 12, 100, 25)
    height = st.number_input('Height (cm)', 100, 250, 175)
    weight = st.number_input('Weight (kg)', 30, 200, 70)
    gender = st.selectbox('Gender', ['Male', 'Female', 'Other'])

    if st.form_submit_button("Generate Workout Plan"):
        try:
            bmi = weight / ((height/100)**2 + 1e-6)
            numerical = scaler.transform([[age, height, weight, bmi]])

            recommendations = []
            for ex_type in ['Aerobic', 'Anaerobic', 'Strength', 'Flexibility']:
                # Encode categoricals
                encoded_cat = encoder.transform([[gender_mapping[gender], ex_type]])

                # Get predictions
                pred = model.predict([numerical, encoded_cat], verbose=0)
                valid_indices = [i for i, name in enumerate(classes) if name in exercise_type_rules[ex_type]]

                if valid_indices:
                    top_idx = valid_indices[np.argmax(pred[0][valid_indices])]
                    workout_name = classes[top_idx]

                    duration = duration_data[
                        (duration_data['exercise_type'] == ex_type) &
                        (duration_data['exercise_name'] == workout_name)
                    ]['exercise_duration'].median()

                    recommendations.append({
                        'type': ex_type,
                        'name': workout_name,
                        'duration': f"{int(duration)} mins" if not pd.isna(duration) else "30-45 mins"
                    })

            st.subheader("Your Personalized Workout Plan")
            cols = st.columns(4)
            for i, rec in enumerate(recommendations):
                with cols[i]:
                    st.markdown(f"""
                    **{rec['type']}**
                    {rec['name']}
                    ⏱️ {rec['duration']}
                    """)

        except Exception as e:
            st.error(f"Error: {str(e)}")
```
